```python
import csv
from models.respostas import db
from models.respostas import Textos, Estudantes, Perguntas, Respostas
```

```python
def get_or_create(session, model, defaults=None, **kwargs):
    instance = session.query(model).filter_by(**kwargs).one_or_none()
    if instance:
        return instance, False
    else:
        kwargs |= defaults or {}
        instance = model(**kwargs)
        try:
            session.add(instance)
            session.commit()
        except Exception:  # The actual exception depends on the specific database so we catch all exceptions. This is similar to the official documentation: https://docs.sqlalchemy.org/en/latest/orm/session_transaction.html
            session.rollback()
            instance = session.query(model).filter_by(**kwargs).one()
            return instance, False
        else:
            return instance, True
```

```python
_resp_path = '../respostas'
_files = ['respostas_ze_gotinha']

_cabecalho = []
respostas = []
for file_ in _files:
    with open(f"{_resp_path}/{file_}.csv") as f:
        _data = csv.reader(f)
        _cabecalho=next(_data)
        respostas = [dict(zip(_cabecalho,i)) for i in _data]
        print(respostas)
```

```python
_texto_ids = {}
_campos = ['Texto',	'ID_Estudante',	'Ano',	'Data_Hora']
for resposta_ in respostas:
    texto_ = resposta_.get('Texto')

    if not texto_:
        continue

    texto_id_ = _texto_ids.get(texto_)
    if not texto_id_:
        texto_obj_, created = get_or_create(db.session, Textos, nome=texto_)
        texto_id_ = texto_obj_.id
        _texto_ids[texto_] = texto_id_

    estudante_id = int(resposta_["ID_Estudante"])
    estudante_obj_, created = get_or_create(db.session, Estudantes, {'ano': resposta_.get('Ano')}, id=estudante_id)

    data_hora_ = resposta_['Data_Hora']

    for k, v in resposta_.items():
        if k not in _campos:
            nr_pergunta = k[0:2]
            txt_pergunta = k[2:]
            pergunta_, created = get_or_create(db.session, Perguntas, {'pergunta': txt_pergunta.strip()}, id=int(nr_pergunta))

            resposta_ = get_or_create(db.session, Respostas,
                                      {'data_hora': data_hora_, 'resposta': v},
                                      texto_id=texto_id_,
                                      estudante_id=estudante_id,
                                      pergunta_id=pergunta_.id,)
            print(resposta_)
```

```python
respostas_objs = Respostas.query.all()

with open('../respostas/ze_gotinha_transform.csv', 'wb') as csvfile:

    fieldnames = [k for k in respostas_objs[0].as_dict.keys()]

    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    for i in respostas_objs:
        writer.writerow(i.as_dict)
```
